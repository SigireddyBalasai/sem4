```python
# working ith multiple arrays
arr1 : np.ndarray = np.array([1,2,3])
arr2 : np.ndarray = np.array([4,5,6])
```

```python
print("addition of array1 and array2",np.add(arr1,arr2))
print("subtraction of array1 and array2",np.subtract(arr1,arr2))
print("Multiplication of array1 and array2",np.multiply(arr1,arr2))
print("dicision of array1 and array2",np.divide(arr1,arr2))
```

```python
print("Min of array",arr1.min())
print("Max of array",arr1.max())
print("sum of array",arr1.sum())
print("Mean of array",arr1.mean())
print("Variance of array",arr1.var())
```
